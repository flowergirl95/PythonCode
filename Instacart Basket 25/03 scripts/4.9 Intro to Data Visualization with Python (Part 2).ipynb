```python
#reduce sample with only columns needed
customer_merged_4= small[['age','income']]
```

```python
#Create a scatterplot
scatterplot= sns.scatterplot( x='age',y='income',data= customer_merged_4)
```

Wide Distribution: The data points are spread across various income levels, indicating that income varies significantly across different age groups. Theres no clear upward or downward trend.

No Strong Correlation: Unlike a linear relationship where income increases or decreases predictably with age, this scatterplot suggests high variability younger and older individuals both fall within various income brackets.
